```python
import pandas as pd

df = pd.read_csv("../data_raw/ecommerce_customer_churn_dataset.csv")
df.head()
```

```python
missing_ratio = df.isnull().mean().sort_values(ascending=False)
missing_ratio
```

```python
missing_by_churn = df.isnull().groupby(df['Churned']).mean().T
missing_by_churn
```

```python
df['Churned'].value_counts(normalize=True)
```

```python
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
numeric_cols = numeric_cols.drop('Churned')

churn_comparison = df.groupby('Churned')[numeric_cols].mean().T
churn_comparison
```

```python
churn_comparison['diff_ratio'] = (
    churn_comparison[1] - churn_comparison[0]
) / churn_comparison[0]

churn_comparison.sort_values('diff_ratio')
```

```python
correlation = df.corr(numeric_only=True)
correlation['Churned'].sort_values(ascending=False)
```

```python
df['recency_bin'] = pd.qcut(df['Days_Since_Last_Purchase'], 5)

df.groupby('recency_bin')['Churned'].mean()
```

```python
df['login_bin'] = pd.qcut(df['Login_Frequency'], 5)

df.groupby('login_bin')['Churned'].mean()
```

```python
df['call_bin'] = pd.qcut(df['Customer_Service_Calls'], 5)
df.groupby('call_bin')['Churned'].mean()
```

```python
df['aov_bin'] = pd.qcut(df['Average_Order_Value'], 5)
df.groupby('aov_bin')['Churned'].mean()
```

```python
df['risk_user'] = (
    (df['Customer_Service_Calls'] >= 5) &
    (df['Days_Since_Last_Purchase'] >= 30)
).astype(int)

df.groupby('risk_user')['Churned'].mean()
```
